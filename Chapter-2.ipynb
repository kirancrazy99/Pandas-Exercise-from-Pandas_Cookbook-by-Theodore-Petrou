### Working with operators on DataFrame

```python
collage = pd.read_csv('D:/python/Pandas/PandasCookbook_Code/data/college.csv',index_col='INSTNM')
collage_ugds = collage.filter(like='UGDS_')
collage_ugds.head()
```

```python
(collage_ugds+.00501).head()
```

```python
((collage_ugds+.00501)//.01).head()
```

```python
collage_ugds_op_round = (collage_ugds+.00501)//.01/100
collage_ugds_op_round.head()
```

```python
collage_ugds_round = (collage_ugds+.0001).round(2)
collage_ugds_round.head()
```

```python
collage_ugds_op_round.head().equals(collage_ugds_round.head())
```

```python
from pandas.testing import assert_frame_equal

assert_frame_equal(collage_ugds_round,collage_ugds_round)
```
