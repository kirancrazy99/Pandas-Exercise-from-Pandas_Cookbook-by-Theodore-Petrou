```python
movie = pd.read_csv('d:/python/Pandas/PandasCookbook_Code/data/movie.csv')
movie.head()
```

```python
movie1=movie.set_index('movie_title')
```

```python
idx_names = {'Avatar':'Ratava','Spectre':'Ertceps'}
column_names = {'director_name':'Director_name','duration':'Duration','num_critic_for_reviews':'Critical_Reviews'}

movie1.rename(index=idx_names,columns=column_names).head()
```

```python
movie1.head()
```

```python
movie1['actor_director_facebook_likes']=movie1['actor_1_facebook_likes']+\
                                        movie1['actor_2_facebook_likes']+\
                                        movie1['actor_3_facebook_likes']+\
                                        movie1['director_facebook_likes']
```

```python
movie1['actor_director_facebook_likes'].isna().sum()
```

```python
movie1['is_cast_likes_more']=movie1['cast_total_facebook_likes']>=movie1['actor_director_facebook_likes']
```

```python
movie1['is_cast_likes_more'].all()
```

```python
movie1 = movie1.drop('actor_director_facebook_likes',axis=1)
```

```python
movie1['actor_total_facebook_likes']=movie1['actor_1_facebook_likes']+\
                                     movie1['actor_2_facebook_likes']+\
                                     movie1['actor_3_facebook_likes']
```

```python
movie1['actor_total_facebook_likes']=movie1['actor_total_facebook_likes'].fillna(0)
```

```python
movie1['is_cast_likes_more']=movie1['cast_total_facebook_likes']>=movie1['actor_total_facebook_likes']
```

```python
movie1['is_cast_likes_more'].all()
```

```python
movie1['pct_actor_cast_like']=movie1['actor_total_facebook_likes']/movie1['cast_total_facebook_likes']
```

```python
movie1['pct_actor_cast_like'].head()
```

```python
profit_index = movie1.columns.get_loc('gross')
```

```python
movie1.insert(loc=profit_index,column='profit',value=movie1['gross']-movie1['budget'])
```
