### Filter rows with Nan values

```python
import numpy as np
technologies= {
    'Courses':["Spark","PySpark","Spark","Python","PySpark","Java"],
    'Fee' :[22000,25000,np.nan,np.nan,np.nan,np.nan],
    'Duration':['30days',np.nan,'30days','N/A', np.nan,np.nan]
          }

df = pd.DataFrame(technologies)
df 
```

### Filter out rows if two or more columns contain Null values

```python
df.dropna(thresh=2)
```

### Filter out only Non nan values in Duration field

```python
df.dropna(subset=['Duration'])
```

### Filter out only Nan values in Duration

```python
df[pd.isnull(df['Duration'])]
```
